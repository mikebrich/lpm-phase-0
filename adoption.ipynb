```python
import pandas as pd
import numpy as np
```

```python
raw_adp_df = pd.read_excel('inputs/PunchSource2023.xlsx') #read in the excel, put into DF
```

```python
raw_adp_df = raw_adp_df[['Last Name', 'First Name', 'Time In']]
```

```python
raw_adp_df
```

```python
raw_adp_df.loc[:, 'Date'] = raw_adp_df['Time In'].dt.date
raw_adp_df.drop(columns=['Time In'], inplace=True)
```

```python
raw_adp_df
```

```python
raw_scheduler_df = pd.read_excel('inputs/SchedulerShifts2023.xlsx') #read in the excel, put into DF
raw_scheduler_df = raw_scheduler_df.drop(columns=["epoch_start_time", "cover_time"]) 
```

```python
#can drop any rows that have cover_type full and role = 2 
# b/c cover_type full either means request in future or was not taken
# 2 = manager so not relevant 
raw_scheduler_df = raw_scheduler_df[~((raw_scheduler_df['role'] == 2.0) | 
                                      (raw_scheduler_df['cover_type'] == 'full'))]

raw_scheduler_df['start_time_est'] = raw_scheduler_df['start_time_est'].dt.date #convert to just date 

raw_scheduler_df
```

```python
start_date = pd.Timestamp('2023-05-25').date()
end_date = pd.Timestamp('2023-08-01').date()
scheduler_filtered_df = raw_scheduler_df[(raw_scheduler_df['start_time_est'] >= start_date) & 
                                         (raw_scheduler_df['start_time_est'] <= end_date)]

scheduler_filtered_df
```

```python
unique_assignee_names = scheduler_filtered_df['assignee_name'].nunique()
unique_assignee_names
```

```python
adp_filtered_df = raw_adp_df[(raw_adp_df['Date'] >= start_date) & 
                             (raw_adp_df['Date'] <= end_date)]

adp_filtered_df = adp_filtered_df.assign( LastFirst = adp_filtered_df['Last Name']+adp_filtered_df['First Name']) 

adp_filtered_df.columns
```
